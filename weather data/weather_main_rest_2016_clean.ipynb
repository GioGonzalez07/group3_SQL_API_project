```python
import pandas as pd
```

```python
weather_main_cities_rest_2016 = pd.read_csv('../data/weather_main_cities_rest_of_2016_202405101142.csv')
```

```python
weather_main_cities_rest_2016.info()
```

```python
weather_main_cities_rest_2016.sample(20)
```

## Drop Columns

```python
weather_main_cities_rest_2016.drop(['weather_wdir', 'weather_wpgt', 'weather_pres', 'weather_tsun'], axis=1, inplace=True)
```

```python
weather_main_cities_rest_2016.info()
```

## Convert format of weather_date Column to datetime

```python
weather_main_cities_rest_2016['weather_date'] = pd.to_datetime(weather_main_cities_rest_2016['weather_date'], format='%Y-%m-%d')
```

Check format again

```python
weather_main_cities_rest_2016.info()
```

## Rename columns

```python
weather_main_cities_rest_2016.rename(columns={
                                    'weather_date': 'date',
                                    'weather_tavg': 'avg_temp_°C',
                                    'weather_tmin': 'min_temp_°C',
                                    'weather_tmax': 'max_temp_°C',
                                    'weather_prcp': 'rainfall_mm',
                                    'weather_snow': 'snowfall_mm',
                                    'weather_wspd': 'avg_windspeed_kmh'
                                    }, inplace=True)
```

```python
weather_main_cities_rest_2016.info()
```

## Create a new column with the names of the cities instead of the code of the station

```python
station_names = {
                    74486: 'New York, NY',
                    72509: 'Boston, MA',	
                    72405: 'Washington, DC',
                    72202: 'Miami, FL',
                    72243: 'Houston, TX',
                    72494: 'San Francisco, CA',
                    72793: 'Seattle, WA'
                }
```

```python
weather_main_cities_rest_2016['city_name'] = weather_main_cities_rest_2016['station'].map(station_names)
```

```python
weather_main_cities_rest_2016.sample(10)
```

```python
weather_main_cities_rest_2016['city_name'].unique()
```

```python
weather_main_cities_rest_2016.drop(['station'], axis=1, inplace=True)
```

```python
weather_main_cities_rest_2016.info()
```

```python
weather_main_cities_rest_2016.sample(20)
```
