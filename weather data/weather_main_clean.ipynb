```python
import pandas as pd
```

```python
weather_main_cities = pd.read_csv('../data/weather_main_cities_202405100846.csv')
```

```python
weather_main_cities.info()
```

```python
weather_main_cities.sample(20)
```

```python
weather_main_cities.duplicated().value_counts()
```

## Drop Columns

```python
weather_main_cities.drop(['weather_wdir', 'weather_wpgt', 'weather_pres', 'weather_tsun'], axis=1, inplace=True)
```

```python
weather_main_cities.info()
```

## Convert format of weather_date Columns to datetime

```python
weather_main_cities['weather_date'] = pd.to_datetime(weather_main_cities['weather_date'], format='%Y-%m-%d')
```

Check the format again

```python
weather_main_cities.info()
```

## Rename columns

```python
weather_main_cities.rename(columns={
                                    'weather_date': 'date',
                                    'weather_tavg': 'avg_temp_°C',
                                    'weather_tmin': 'min_temp_°C',
                                    'weather_tmax': 'max_temp_°C',
                                    'weather_prcp': 'rainfall_mm',
                                    'weather_snow': 'snowfall_mm',
                                    'weather_wspd': 'avg_windspeed_kmh'
                                    }, inplace=True)
```

```python
weather_main_cities.info()
```

## Drop rows of city=New Jersey, because we decided not to use it

```python
weather_main_cities = weather_main_cities[weather_main_cities['station'] != 72502]
```

```python
# Check if New Jersey really dropped
weather_main_cities['station'].unique()
```

## Create a new column with the names of the cities instead of the code of the station:

```python
station_names = {
                    74486: 'New York, NY',
                    72509: 'Boston, MA',	
                    72405: 'Washington, DC',
                    72202: 'Miami, FL',
                    72243: 'Houston, TX',
                    72494: 'San Francisco, CA',
                    72793: 'Seattle, WA'
                }
```

```python
weather_main_cities['city_name'] = weather_main_cities['station'].map(station_names)
```

```python
weather_main_cities.sample(10)
```

```python
weather_main_cities['city_name'].unique()
```

```python
weather_main_cities.drop(['station'], axis=1, inplace=True)
```

```python
weather_main_cities.info()
```

```python
weather_main_cities.sample(20)
```
